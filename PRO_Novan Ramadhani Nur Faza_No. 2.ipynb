```python
import tkinter as tk
from sympy import*
from tkinter import*
```

```python
def bangun():
    if rr.get()==1:
        kubus = Toplevel(root)
        kubus.title('Kubus')
        kubus.geometry("400x400")
        Label_1 = tk.Label(kubus, text="Masukan Sisi :").pack()
        entry_1 = tk.Entry(kubus, textvariable = pk).pack()
        res = pk.get()**3
        myText.set(res)
        res_button = tk.Button(kubus,text="Hitung",
                      command=hitung).pack()
        result = tk.Label(kubus, text="(Result)", textvariable = myText).pack()
        exit_button = tk.Button(kubus,text="Exit",
                           command=kubus.destroy).pack()
        
    elif rr.get()==2:
        balok = Toplevel(root)
        balok.title('Balok')
        balok.geometry("400x400")
        Label_1 = tk.Label(balok, text="Masukan Panjang :").pack()
        entry_1 = tk.Entry(balok, textvariable = pb).pack()
        Label_2 = tk.Label(balok, text="Masukan Lebar :").pack()
        entry_2 = tk.Entry(balok, textvariable = qb).pack()
        Label_3 = tk.Label(balok, text="Masukan Tinggi :").pack()
        entry_3 = tk.Entry(balok, textvariable = rb).pack()
        res = pb.get()*qb.get()*rb.get()
        myText.set(res)
        result = tk.Label(balok, text="(Result)", textvariable = myText).pack()
        res_button = tk.Button(balok,text="Hitung",
                      command=hitung).pack()
        exit_button = tk.Button(balok,text="Exit",
                           command=balok.destroy).pack()

    elif rr.get()==3:
        bola = Toplevel(root)
        bola.title('Bola')
        bola.geometry("400x400")
        Label_1 = tk.Label(bola, text="Masukan Jari-jari :").pack()
        entry_1 = tk.Entry(bola, textvariable = pl).pack()
        res = (pl.get()**3)*(4/3)*3.14
        myText.set(res)
        res_button = tk.Button(bola,text="Hitung",
                      command=hitung).pack()
        result = tk.Label(bola, text="(Result)", textvariable = myText).pack()
        exit_button = tk.Button(bola,text="Exit",
                           command=bola.destroy).pack()
  
    elif rr.get()==4:
        tabung = Toplevel(root)
        tabung.title('Tabung')
        tabung.geometry("400x400")
        Label_1 = tk.Label(tabung, text="Masukan Jari-jari :").pack()
        entry_1 = tk.Entry(tabung, textvariable = pt).pack()
        Label_2 = tk.Label(tabung, text="Masukan Tinggi :").pack()
        entry_2 = tk.Entry(tabung, textvariable = qt).pack()
        res = (pt.get()**2)*qt.get()*3.14
        myText.set(res)
        res_button = tk.Button(tabung,text="Hitung",
                      command=hitung).pack()
        result = tk.Label(tabung, text="(Result)", textvariable = myText).pack()
        exit_button = tk.Button(tabung,text="Exit",
                           command=tabung.destroy).pack()
    
    else:
        limas = Toplevel(root)
        limas.title('Limas Persegi Panjang')
        limas.geometry("400x400")
        Label_1 = tk.Label(limas, text="Masukan Panjang :").pack()    
        entry_1 = tk.Entry(limas, textvariable = pm).pack()
        Label_2 = tk.Label(limas, text="Masukan Lebar :").pack()
        entry_2 = tk.Entry(limas, textvariable = qm).pack()
        Label_3 = tk.Label(limas, text="Masukan Tinggi :").pack()
        entry_3 = tk.Entry(limas, textvariable = rm).pack()
        res = pm.get()*qm.get()*rm.get()*1/3
        myText.set(res)
        res_button = tk.Button(limas,text="Hitung",
                      command=hitung).pack()
        result = tk.Label(limas, text="(Result)", textvariable = myText).pack()
        exit_button = tk.Button(limas,text="Exit",
                           command=limas.destroy).pack()
def hitung():
    if rr.get()==1:
        res = pk.get()**3
    elif rr.get()==2:
        res = pb.get()*qb.get()*rb.get()
    elif rr.get()==3:
        res = (pl.get()**3)*(4/3)*3.14
    elif rr.get()==4:
        res = (pt.get()**2)*qt.get()*3.14
    else:
        res = pm.get()*qm.get()*rm.get()*1/3
    myText.set(res)
```

```python
root = tk.Tk()
root.title("Menghitung Volume Bangun Ruang")
root.geometry("400x400")

myText = tk.StringVar()
pk = tk.DoubleVar()
pb = tk.DoubleVar()
qb = tk.DoubleVar()
rb = tk.DoubleVar()
pl = tk.DoubleVar()
pt = tk.DoubleVar()
qt = tk.DoubleVar()
pm = tk.DoubleVar()
qm = tk.DoubleVar()
rm = tk.DoubleVar()
rr = tk.IntVar()


Label_1 = tk.Label(root, text="Pilih Jenis Bangun Ruang").pack()

button_1 = tk.Radiobutton(root, text="Kubus", variable = rr, value = 1).pack()
button_2 = tk.Radiobutton(root, text="Balok", variable = rr, value = 2).pack()
button_3 = tk.Radiobutton(root, text="Bola", variable = rr, value = 3).pack()
button_4 = tk.Radiobutton(root, text="Tabung", variable = rr, value = 4).pack()
button_5 = tk.Radiobutton(root, text="Limas Persegi Panjang", variable = rr, value = 5).pack()

button = tk.Button(root,text="Pilih", command=bangun).pack()
exit_button = tk.Button(root,text="Exit", command=root.destroy).pack()
root.mainloop()
```
